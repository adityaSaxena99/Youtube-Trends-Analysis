Next up, We're going to Add a new column to the datasets that signify what country the data comes from. This will be important since we will then create a dataset that has all 5 datasets combined.

```python
USAdata['Country']= 'United States'
GBdata['Country']= 'Great Britain'
CAdata['Country']= 'Canada'
JPdata['Country']= 'Japan'
KRdata['Country']= 'Korea'
```

```python
Alldata = pd.concat([USAdata, GBdata, CAdata, JPdata, KRdata])
```

```python
Alldata
```

First Lets find the total count of trending videos per country and bind it to a Variable.

```python
count_of_videos = Alldata.value_counts('Country')
count_of_videos
```

As you can see, The data for the Variable count_of_videos shows us the total count of trending videos per country.
Now lets make some visual diagrams of this data.

```python
convalues = [count_of_videos[0], count_of_videos[1], count_of_videos[2], count_of_videos[3], count_of_videos[4]]
widths = [0.7, 0.7, 0.7, 0.7, 0.7]
colors = ['k', 'g', 'b', 'c', 'r']

fig, ax= plt.subplots()

plt.bar(range(5), convalues, width=widths, color=colors, align='center')
plt.xticks(range(5), ('United States', 'Canada', 'Great Britain', 'Korea', 'Japan'))
ax.set_ylabel('Count of Trending Videos')
plt.title('The Amount of Trending videos per Country')

plt.show()
```

```python
convalues = [count_of_videos[0], count_of_videos[1], count_of_videos[2], count_of_videos[3], count_of_videos[4]]
colors = ['m', 'g', 'b', 'c', 'r']
labels = ['United States', 'Canada', 'Great Britain', 'Korea', 'Japan']

plt.pie(convalues, colors=colors, labels=labels, autopct= '%.1f%%', counterclock= False, shadow= True)
plt.title('The Amount of Trending videos per Country')
plt.show
```

From the Data we have visualized we know that in terms of most trending videos 1st: United States, 2nd: Canada, 3rd: Great Britain, 4th: Korea, 5th: Japan. We also know from the pie graph that when summing the total videos togther, we know that the US had 23.3% of the total videos (etc.)
